```python
import numpy as np
import scipy as sp
import scipy.signal as spsig
import matplotlib.pyplot as plt
```

```python
class Sig:
    '''Класс сигнала с возможностью включения и выключения импульсных помех'''
    
    i1 = np.complex(0.0, 1.0)
    
    def __init__(self, dt, N, Amp_noise, Amp_target, f1):
        self.dt = dt
        self.N = N
        self.Amp_noise = Amp_noise
        self.Amp_target = Amp_target
        self.f1 = f1
        self.Signal = np.zeros(N, dtype=np.complex)
        for n in range(N):
            self.Signal[n] = A3 * np.exp(-self.i1*2.0*np.pi*f1*dt*(n-1)) + Amp_noise*1.0/np.sqrt(2.0)* (np.random.randn()+self.i1*np.random.randn());
        
    def AddBregg(self, A1, A2, f_br):
        self.A1 = A1
        self.A2 = A2
        self.f_br = f_br
        self.Bregg = np.zeros(self.N, dtype=np.complex)
        for n in range(N):
            self.Bregg[n] = A1 * np.exp(-self.i1*2.0*np.pi*(-f_br)*dt*(n-1)) + A2*np.exp(-self.i1*2.0*np.pi*f_br*self.dt*(n-1))
        
        self.Signal = self.Signal + self.Bregg
        
    def AddImpClutter(self, xls, xrs, As):
        self.ImpClutter = np.zeros(self.N,dtype=np.complex)
        for i in range(len(As)):
            for n in range(xls[i],xrs[i]):
                self.ImpClutter[n] = As[i]*1.0/np.sqrt(2.0)*(np.random.randn()+self.i1*np.random.randn())
             
        self.Signal = self.Signal + self.ImpClutter
            
    def Plot(self):
        plt.plot(np.abs(self.Signal))
        plt.grid()
        plt.title('Main Signal');
    
    def PlotBregg(self):
        plt.plot(np.abs(self.Bregg))
        plt.grid()
        plt.title('Main Breggl');
        
    def Spector(self):
        self.w = spsig.chebwin(self.N, 80)
        self.sp = np.fft.fft(self.w * self.Signal)
            
    def PlotSpector(self):
        self.w = spsig.chebwin(self.N, 80)
        self.sp = np.fft.fftshift(np.fft.fft(self.w * self.Signal))
        plt.plot(np.abs(self.sp))
        plt.grid()
        plt.title('Fourier Main Signal');
    
    def AddZeros(self, xls, xrs):
        for i in range(len(xls)):
            for n in range(xls[i],xrs[i]):
                self.Signal[n] = np.complex(0.0, 0.0)
    
    def Filter(self, coeffs):
        Nv = len(coeffs)
        self.FilteredSignal = np.zeros(len(self.Signal), dtype=np.complex)
        self.tmp = np.array([np.dot(self.Signal[n:n+Nv], coeffs) for n in range(N-Nv)],dtype=np.complex)       
        self.FilteredSignal[:len(self.tmp)] = self.tmp
    
    def PlotFilter(self):
        plt.plot(np.abs(self.FilteredSignal))
        plt.grid()
        plt.title('Main Signal Filtered');
        
    def PlotFilterSpector(self):
        self.w = spsig.chebwin(self.N, 80)
        self.sp = np.fft.fftshift(np.fft.fft(self.w * self.FilteredSignal))
        plt.plot(np.abs(self.sp))
        plt.grid()
        plt.title('Fourier Main Signal');
        
```

```python
def rotate(A):
    N = A.shape[0]
    B = np.zeros((N,N),dtype=np.complex)
    for i in range(N):
        for j in range(N):
            B[i,j] = A[i,N-j-1]
    return B
```
